# 2x + 3y = 8
# 5x + 4y = 20

```python
# coefficient Matrix
A = np.array([[2,3],[5,4]]) # 2x2 matrxi
A
```

```python
# contant matrix
B = np.array([8,20])
B
```

```python
X = A/B
X
```

```python
# solve linear algebra
X = np.linalg.solve(A,B)
X
```

# Vector Pin pointing

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v = np.array([2,3])
v2 = np.array([3,4])
v3 = np.array([4,5])
```

```python
plt.figure(figsize=(4,4))
plt.quiver(0,0,v[0],v[1], angles = 'xy', scale_units = 'xy',scale =1, color = 'g')
plt.quiver(0,0,v2[0],v2[1], angles = 'xy', scale_units = 'xy',scale =1, color = 'b')
plt.quiver(0,0,v3[0],v3[1], angles = 'xy', scale_units = 'xy',scale =1, color = 'r')
plt.xlim(-0,5)
plt.ylim(-0,5)
plt.grid()
plt.title('2D Vector')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.gca().set_aspect('equal',adjustable='box')
plt.show()
```

```python
# Linear transformation of the matrix
# define rotation matrix of 45
theta = np.radians(30)
rotation_mat = np.array([[np.cos(theta),-np.sin(theta)],
                       [np.sin(theta), np.cos(theta)]])
# APPLY TRANSFORMATION
v_trans = np.dot(rotation_mat,v)
v_trans
plt.figure(figsize=(6,6))
plt.quiver(0,0,v[0],v[1], angles = 'xy', scale_units = 'xy',scale =1, color = 'r', label = "original")
plt.quiver(0,0,v_trans[0],v_trans[1], angles = 'xy', scale_units = 'xy',scale =1, color = 'b', label = 'Transformed')
plt.xlim(-1,5)
plt.ylim(-1,5)
plt.grid()
plt.title('2D Vector')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.gca().set_aspect('equal',adjustable='box')
plt.show()
```

```python
# Sheear Transformation Define shear factor
shear_factor = 1.5
shear_matrix = np.array([[1, shear_factor], [0, 1]])
# apply shear transformation
v_sheared = np.dot(shear_matrix,v)
# plotting original rotated and sheared 
plt.figure(figsize=(6,6))
plt.quiver(0,0,v[0],v[1], angles = 'xy', scale_units = 'xy',scale =1, color = 'r', label = "original")
plt.quiver(0,0,v_trans[0],v_trans[1], angles = 'xy', scale_units = 'xy',scale =1, color = 'b', label = 'Transformed')
plt.quiver(0,0,v_sheared[0],v_sheared[1], angles = 'xy', scale_units = 'xy',scale =1, color = 'g', label = 'shear_Transformed')
plt.xlim(-1,7)
plt.ylim(-1,7)
plt.grid(True, which='both', linestyle = '--',linewidth = 0.5)
plt.axhline(y=0, color='k', linewidth = 0.5)
plt.axvline(x=0, color='k', linewidth = 0.5)
plt.title('2D Vector')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.gca().set_aspect('equal',adjustable='box')
plt.show()
```

```python
# adjust by including the effect of shearing on the grid
# create gird of points
x = np.linspace(-5, 10, 10)
y = np.linspace(-5, 10, 10)
X,Y = np.meshgrid(x,y)
# applying shear transformation tot he grid points
# X_sheared, Y_sheared = np.vstack(shear_factor, [X.ravel(), Y.ravel()])
# X_sheared  = X_sheared.reshape(X.shape)
# Y_sheared = Y_sheared.reshape(Y.shape)
grid_points = np.vstack([X.ravel(), Y.ravel()])
sheared_points = np.dot(shear_matrix, grid_points)
X_sheared = sheared_points[0].reshape(X.shape)
Y_sheared = sheared_points[1].reshape(Y.shape)
# plotting original rotated and sheared 
plt.figure(figsize=(6,6))
plt.quiver(0,0,v[0],v[1], angles = 'xy', scale_units = 'xy',scale =1, color = 'r', label = "Original")
plt.quiver(0,0,v_trans[0],v_trans[1], angles = 'xy', scale_units = 'xy',scale =1, color = 'b', label = 'Transformed')
plt.quiver(0,0,v_sheared[0],v_sheared[1], angles = 'xy', scale_units = 'xy',scale =1, color = 'g', label = 'shear_Transformed')

for i in range(len(x)):
    plt.plot(X_sheared[i, :], Y_sheared[i, :],  'k--', linewidth =0.5)
    plt.plot(X_sheared[:, i], Y_sheared[:, i], 'k--', linewidth = 0.5)


plt.xlim(-1,7)
plt.ylim(-1,7)
plt.grid(True, which='both', linestyle = '--',linewidth = 0.5)
plt.axhline(y=0, color='r', linewidth = 1)
plt.axvline(x=0, color='b', linewidth = 1)
plt.title('2D Vector')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.gca().set_aspect('equal',adjustable='box')
plt.show()
```

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# load data from sns
df = sns.load_dataset('tips')
df.columns
```

```python
# Draw a Histplot with 
plt.figure(figsize=(8, 6))
sns.histplot(df['total_bill'], kde=True)
```

```python
# Draw a Histplot with 
plt.figure(figsize=(8, 6))
sns.histplot(df['tip'], kde=True)
```

```python
# Draw a Histplot with 
plt.figure(figsize=(8, 6))
sns.histplot(df['day'], kde=True)
sns.histplot(df['size'], kde=True)
```

```python
# Normal data
normal_data = np.random.normal(size = 1000, loc = 0, scale = 1)
# Uniform data
uniform_data = np.random.uniform(low = -2, high = 2, size=1000)
# exponential data
exponential_data = np.random.exponential(scale = 1, size = 1000)
plt.figure(figsize=(12,8))
# normal sitribution
plt.subplot(3,1,1)
sns.histplot(normal_data, bins=30, color= 'lightblue', alpha = 0.8, kde= True)
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
# uniform distributions
plt.subplot(3,1,2)
sns.histplot(uniform_data, bins=30, color= 'lightgreen', alpha = 0.8, kde= True)
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Exoponential distribution
plt.subplot(3,1,3)
sns.histplot(exponential_data, bins=30, color= 'salmon', alpha = 0.8, kde= True)
plt.title('Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')


plt.tight_layout()
plt.show()
```

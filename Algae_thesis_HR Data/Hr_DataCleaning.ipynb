### Display Summay and Info of DataSet

```python
#Find Unique Values and replace empty spaces of data set with underscore
df["Business Travel"].unique()
df.columns = df.columns.str.replace(" ", "_")
```

```python
df["CF_age_band"].unique()
cat_map = {
    "Under 25": "X1",
    "25 - 34": "X2",
    "35 - 44": "X3",
    "45 - 54": "X4",
    "Over 55": "X5"
}
df["CF_age_band"] = df['CF_age_band'].map(cat_map)
```

```python

# Create a scatter plot with customization
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,  # Pass the DataFrame here
    x= 'Num_Companies_Worked',  # X-axis
    y='Age',  # Y-axis
    hue='CF_age_band',  # Color by age band
    style='CF_age_band',  # Different marker styles for each category
    s=100,  # Marker size
    edgecolor='black',  # Edge color of markers
    alpha=0.7  # Marker transparency
)

# Customize the plot
plt.title('Scatter Plot of Number of Companies Worked vs. Age', fontsize=16)
plt.xlabel('Number of Companies Worked', fontsize=14)
plt.ylabel('Age', fontsize=14)
plt.legend(title='Age Band', title_fontsize='13', loc='upper left')

# Show plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.io as pio
from plotly.subplots import make_subplots
from plotly.graph_objects import Heatmap
from sklearn.preprocessing import MinMaxScaler

# Load the data from the Excel file
file_path = "D:/Study/Study and Extras/Dr Nawaz Sb Paper/Algae Paper Abstract Rehan Islamabad/Donut and Heatmap data.xlsx"
df = pd.read_excel(file_path, sheet_name="Mean_value")

# Extract unique values of days
unique_days = df['days'].unique()

# Create a colormap with multiple colors
cmap = plt.get_cmap("tab20")
colors = [cmap(i) for i in range(cmap.N)]
colorscale = [(i / (cmap.N - 1), f'rgb({int(r * 255)}, {int(g * 255)}, {int(b * 255)})') for i, (r, g, b, _) in enumerate(colors)]

# Define a mapping for day labels
day_labels = {
    'Day7': 'Day 7',
    'Day14': 'Day 14',
    'Day21': 'Day 21',
    'Day28': 'Day 28'
}

# Create a subplot figure
fig = make_subplots(rows=2, cols=2, subplot_titles=[day_labels.get(day, day) for day in unique_days])

# Plot heatmap for each unique day separately
for i, day in enumerate(unique_days):
    day_data = df[df['days'] == day]
    
    # Set the variables as the x-axis and the first column as the y-axis
    x = day_data.columns[2:20]
    y = day_data.iloc[:, 1]
    
    # Convert DataFrame to matrix
    data_matrix = day_data.iloc[:, 2:20].values
    
    # Normalize the data_matrix
    scaler = MinMaxScaler()
    data_matrix_normalized = scaler.fit_transform(data_matrix)
    
    # Create the heatmap using plotly.graph_objects
    heatmap = Heatmap(z=data_matrix_normalized, x=x, y=y, colorscale=colorscale, colorbar=dict(title='Value'))
    
    # Add the heatmap to the subplot
    fig.add_trace(heatmap, row=(i // 2) + 1, col=(i % 2) + 1)

# Update layout
fig.update_layout(height=800, width=1200, title_text="Heatmaps for Different Days", showlegend=False)

# Save the figure as SVG
pio.write_image(fig, "heatmap_combined.svg", format="svg")

# Show the figure
fig.show()
```

```python
# import libraries
import pandas  as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# ML librarries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder, MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# import data
df = sns.load_dataset('tips')
```

```python
# separate the features X and labels/targets Y
X = df[['total_bill']] # X should always be double simension. use double [[]]
y = df['tip']
```

```python
# before train and test scalling
# scaler = StandardScaler()
# X = scaler.fit_transform(X)
```

```python
# train test split
X_train, X_test, y_train, y_test =  train_test_split(X,y, test_size=0.2, train_size=0.8)
```

```python
# # proprocess the data before and after train and test, but ammar preffer after
# scaler = StandardScaler()
# X_train = scaler.fit_transform(X_train)
# X_test = scaler.transform(X_test)
# y_train = scaler.fit_transform(y_train.values.reshape(-1,1))
# y_test = scaler.transform(y_test.values.reshape(-1,1))
```

```python
#Build the Model and call it
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# predict the values with model
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
model.predict([[80]])
```

```python
# Evaluate the mode
print('MSE:', mean_squared_error(y_test, y_pred))
print('R2:', r2_score(y_test, y_pred))
```

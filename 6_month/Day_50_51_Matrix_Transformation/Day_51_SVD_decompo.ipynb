# SVD in `Python`

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(7)
# import random as rd
A =  np.random.rand(4,3)
A
```

```python
# lets apply svd
U, sigma, VT = np.linalg.svd(A)
print("U Matrix=",U)
print("SigmaU Matrix=", sigma)
print("VT Matrix=", VT)
```

```python
print(U.shape)
print(sigma.shape)
print(VT.shape)
```

# Cross Validation

```python
%%time
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB

# load iris data
iris = load_iris()
# define Gaussian Naive bayes model 
nb = GaussianNB()

# Perform K-Fold Cross Validation with K = 5
scores = cross_val_score(nb, iris.data, iris.target, cv = 5, scoring='accuracy')
# print the score for each fold and then mean score
print("Socres for each fold", scores)
print("Socres for each fold", scores.mean())
print("Socres for each fold", scores.std())
```

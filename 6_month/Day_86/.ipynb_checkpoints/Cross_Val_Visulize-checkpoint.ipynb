```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
# Load Iris Data
df = load_iris()
X = df.data
y = df.target
# define decision tree
model = DecisionTreeClassifier()
# perform K  fold Cros
score = cross_val_score(model, X,y, cv=8)
mean_score = np.mean(score)
# plot the figure
plt.figure(figsize=(8,6))
plt.plot(range(1,9), score, color='blue', linestyle = 'dashed', marker = 'o')
# Add a red dot for the mean accuracy
plt.scatter(5.5, mean_score, color='red', s=200, label=f'Mean Accuracy: {mean_score:.2f}')
plt.title(' K-Fold Cross-Validation Results')
plt.xlabel('Fold Number')
plt.ylabel('Accuracy')
plt.ylim([0.5,1.5])
plt.show()
```

# Ensamble Algorythems
- Combine Multiple machine learning models in order to predict best 

- Types to Ensamble
    - Bagging (Bootstrap Aggregating)
        - reduce variance, prevent overfitting
        - create multiple subsets
        - train model on each subset
        - e.g. Randomforetalgorithm
    - Boosting
        - Turning weak learner into strong ones iteratively
        - adjust weight on basis of last classification
        - if observation classified incorrectly, it will try to increase the weight of observation and vice versaw
        - e.g. AdaBoost,Gradient Boosting, XGBoost
    - stacking(Stacked Generalized)
        - stack and train different methods like
            - Logistic Regression + Decision Tree + SVM
- Why this Ensambling
    - High accuracy
    - Stability
    - Reduced Overfitting
- Where used
    - in finance, credit scoring and algorithems
    - Healtcare
    - e-commerce
- Challanges
    - complexity
    - interpretabely
    - parameter tunning

```python
# import libararies
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# load the data
df = sns.load_dataset('tips')
df.head()
```

```python
# encode features
le = LabelEncoder()
for col in df.select_dtypes(include=['category', 'object']):
    df[col] = le.fit_transform(df[col])
df.head()
```

```python
# Split data
X = df.drop(['sex'], axis = 1)
y = df['sex']
# train test
X_train, X_test, y_train, y_test =  train_test_split(X,y, test_size=0.2, train_size=0.8, random_state=42)
```

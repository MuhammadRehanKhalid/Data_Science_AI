```python
sns.scatterplot(data = df, x = 'fare', y = 'age', hue = 'class')
df.plot(kind = 'scatter', y = 'fare', x = 'age', c = 'class', colormap = 'viridis')
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame
df.plot(kind='scatter', x='fare', y='age', c='class', colormap='viridis')

# Add labels and legend
plt.xlabel('Fare')
plt.ylabel('Age')
plt.legend(title='Class')

# Show the plot
# plt.show()
```

```python
# make the above plots using plotly
import plotly.express as px

# Assuming 'df' is your DataFrame
fig = px.scatter(df, x='age', y='fare', color='class')

# Add labels and legend
fig.update_layout(xaxis_title='Age', yaxis_title='Fare', legend_title='Class')

# Show the plot
fig.show()
```

```python
import plotly.graph_objects as go

import pandas as pd

# Read data from a csv
z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')

fig = go.Figure(data=[go.Surface(z=z_data.values)])

fig.update_layout(title='Mt Bruno Elevation', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```

```python
# Normal data
normal_data = np.random.normal(size = 1000, loc = 0, scale = 1)
# Uniform data
uniform_data = np.random.uniform(low = -2, high = 2, size=1000)
# exponential data
exponential_data = np.random.exponential(scale = 1, size = 1000)
plt.figure(figsize=(12,8))
# normal sitribution
plt.subplot(3,1,1)
sns.histplot(normal_data, bins=30, color= 'cyan', alpha = 0.8, kde= True)
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
# uniform distributions
plt.subplot(3,1,2)
sns.histplot(uniform_data, bins=30, color= '#7209B7', alpha = 0.8, kde= True)
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Exoponential distribution
plt.subplot(3,1,3)
sns.histplot(exponential_data, bins=30, color= 'salmon', alpha = 0.8, kde= True)
plt.title('Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')


plt.tight_layout()
plt.show()
```

### Skewness ans Kurtosis

```python
# kurtosis with higer peak called leptokurtic
# kutsosis with plain peak called platokurtic
# normal peak called mesokurtic
```

```python
print(stats.kurtosis(df1['fare']))
print(stats.skew(df1['fare']))
sns.histplot(df1['fare'], kde = True)
```

```python
# generate normal data
np.random.seed(0)
dff = np.random.normal(1,2,1000)
print(stats.kurtosis(dff))
print(stats.skew(dff))
sns.histplot(dff, kde = True, color = 'salmon')
```

```python
df_first = df1[df1['class']== 'First']
print(stats.kurtosis(df_first['fare']))
print(stats.skew(df_first['fare']))
sns.histplot(df_first['fare'], kde = True, color='cyan')
```

```python
df_sec = df1[df1['class']== 'Second']
print(stats.kurtosis(df_sec['fare']))
print(stats.skew(df_sec['fare']))
sns.histplot(df_sec['fare'], kde = True, color='purple')
```

```python
df_thir = df1[df1['class']== 'Third']
print(stats.kurtosis(df_thir['fare']))
print(stats.skew(df_thir['fare']))
sns.histplot(df_thir['fare'], kde = True, color= 'red')
```

```python
# How to test normal distriubuton of data
# H0, H1 normal and not normal
# Shapiro Wilk
# D'Agostino's K^2 Test
# Anderson-Darling test
# Kolmogorov-Smirnov test
```

```python
# Shapiro Wilk Test
stats.shapiro(dff)
# print statistics and P value separately
stat, p = stats.shapiro(dff)
print('Statistics=%.3f, p=%.3f' % (stat, p))
```

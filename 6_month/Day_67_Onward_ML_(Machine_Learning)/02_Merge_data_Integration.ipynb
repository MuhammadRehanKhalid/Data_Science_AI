```python
import pandas as pd
# Example dataset 
data1 = {
    'id':[1,2,3,4],
    'name': ['ahmad', 'fatima','rehan','hassan'],
    'age': [12,23,39,21]
}
df1 = pd.DataFrame(data1)
df1
```

```python
# Example dataset 
data2 = {
    'id':[1,2,3,5],
    'city': ['Karachi', 'Lahore','Faisalabad','Islamabad'],
    'occupation': ['engineer','carpenter','datascientist','data analyst']
}
df2 = pd.DataFrame(data2)
df2
```

```python
df3 = pd.merge(df1, df2, on='id', how='inner')
df4 = pd.merge(df1, df2, on='id', how='outer')
df5 = pd.merge(df1, df2, on='id', how='left')
df6 = pd.merge(df1, df2, on='id', how='right')
df7 = pd.merge(df1, df2, how='cross')
df3
df4
df5
df6
df7
```

```python
import math
```

```python
# Example dataset 
# lets say we have a dataset with two classes , A and B
# Suppose in a data set of 10 elements , 4 are of class A dn 6 are of class B

df_a = 4
df_b = 6
total = df_a+df_b
```

```python
# calculat the proportions
p_a = df_a/total
p_b = df_b/total
p_a
```

```python
# Entropy
entropy = -p_a * math.log2(p_a) - p_b * math.log2(p_b)
print('Entropy: ', entropy)
```

```python
# Gini impurity 
gini = 1-(p_a**2+p_b**2)
print('Gini: ',gini)
```

```python
# Custom Horizontal Percentage Comparison Graphs with Smart Design and Bright Colors
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set style for professional plots
plt.style.use('default')
sns.set_palette("bright")

print("=== HORIZONTAL PERCENTAGE ANALYSIS WITH SMART DESIGN & BRIGHT COLORS ===")
print(f"Dataset shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
print("\nData overview:")
print(df.head())

def get_significance_stars(p_value):
    """Return significance stars based on p-value"""
    if pd.isna(p_value):
        return ""
    elif p_value < 0.001:
        return "***"
    elif p_value < 0.01:
        return "**"
    elif p_value < 0.05:
        return "*"
    else:
        return "ns"

# Create horizontal individual graphs for each row with smart design
def create_horizontal_percentage_graphs_with_p_column(df):
    """Create separate horizontal percentage comparison graph for each row using smart bright design"""
    
    # Define BRIGHT and VIBRANT color scheme for ALL graphs
    # More saturated and eye-catching colors
    consistent_colors = ['#FF0066', '#00CCFF']  # Bright Pink and Bright Cyan
    
    # Alternative bright color schemes for variety
    bright_color_schemes = [
        ['#FF0066', '#00CCFF'],  # Bright Pink & Cyan
        ['#FF3300', '#00FF66'],  # Bright Red & Green  
        ['#FF6600', '#0066FF'],  # Bright Orange & Blue
        ['#CC00FF', '#FFCC00'],  # Bright Purple & Yellow
        ['#FF0099', '#00FF99'],  # Bright Magenta & Mint
    ]
    
    results_summary = []
    
    # Find the required columns
    responsive_col = None
    choice_a_col = None
    choice_b_col = None
    p_values_col = None
    
    # Look for columns by name patterns
    for col in df.columns:
        if 'responsive' in col.lower():
            responsive_col = col
        elif 'choice' in col.lower() and 'a' in col.lower():
            choice_a_col = col
        elif 'choice' in col.lower() and 'b' in col.lower():
            choice_b_col = col
        elif 'p' in col.lower() and 'value' in col.lower():
            p_values_col = col
    
    print(f"Found columns:")
    print(f"  Responsive: {responsive_col}")
    print(f"  Choice A: {choice_a_col}")
    print(f"  Choice B: {choice_b_col}")
    print(f"  P Values: {p_values_col}")
    
    if not all([responsive_col, choice_a_col, choice_b_col]):
        print("‚ùå Error: Could not find required columns (Responsive, Choice A, Choice B)")
        return pd.DataFrame()
    
    # Special handling for merging rows 3 and 4 (indices 2 and 3)
    merge_rows_34 = True
    processed_rows = set()
    
    for idx, row in df.iterrows():
        # Skip row 4 (index 3) if we're merging it with row 3 (index 2)
        if merge_rows_34 and idx == 3:
            continue
            
        # Extract data for current row
        row_data = {
            'row_index': idx,
            'row_name': f"Experiment_{idx+1}" if not (merge_rows_34 and idx == 2) else "Experiment_3_and_4_Combined"
        }
        
        # Extract actual values from columns
        try:
            if merge_rows_34 and idx == 2:
                # Combine data from rows 3 and 4 (indices 2 and 3)
                row3 = df.iloc[2]
                row4 = df.iloc[3]
                
                # Sum the values from both rows
                total_responsive = row3[responsive_col] + row4[responsive_col]
                n_choice_a = row3[choice_a_col] + row4[choice_a_col]
                n_choice_b = row3[choice_b_col] + row4[choice_b_col]
                
                # Use p-value from row 3 (or could average them)
                p_value = row3[p_values_col] if p_values_col else np.nan
                
                # Extract labels from row 3's first column
                first_col_value = str(row3.iloc[0])
                
            else:
                # Regular single row processing
                total_responsive = row[responsive_col]
                n_choice_a = row[choice_a_col]
                n_choice_b = row[choice_b_col]
                p_value = row[p_values_col] if p_values_col else np.nan
                first_col_value = str(row.iloc[0])
            
            # Calculate percentages
            percent_a = (n_choice_a / total_responsive) * 100 if total_responsive > 0 else 0
            percent_b = (n_choice_b / total_responsive) * 100 if total_responsive > 0 else 0
            
            significance = get_significance_stars(p_value)
            
            # Extract labels from first column (assuming it contains comparison like "A vs B")
            if 'vs' in first_col_value.lower():
                # Split by 'vs' and clean up the labels
                label_parts = first_col_value.split('vs')
                if len(label_parts) >= 2:
                    choice_a_label = label_parts[0].strip()
                    choice_b_label = label_parts[1].strip()
                else:
                    choice_a_label = "Choice A"
                    choice_b_label = "Choice B"
            else:
                # Fallback if no 'vs' found
                choice_a_label = "Choice A"
                choice_b_label = "Choice B"
            
            row_data.update({
                'total_responsive': total_responsive,
                'n_choice_a': n_choice_a,
                'n_choice_b': n_choice_b,
                'percent_a': percent_a,
                'percent_b': percent_b,
                'p_value': p_value,
                'significance': significance,
                'choice_a_label': choice_a_label,
                'choice_b_label': choice_b_label
            })
            
        except Exception as e:
            print(f"Error processing row {idx}: {e}")
            continue
        
        # Create individual HORIZONTAL plot for this row - THINNER and SMARTER
        fig, ax = plt.subplots(1, 1, figsize=(9, 3.5))  # Even more compact!
        
        # Use BRIGHT colors with variation for visual interest
        color_scheme = bright_color_schemes[idx % len(bright_color_schemes)]
        
        # Data for plotting (HORIZONTAL) - Use extracted labels
        categories = [choice_a_label, choice_b_label]
        percentages = [percent_a, percent_b]
        sample_info = [f'n={int(n_choice_a)}', f'n={int(n_choice_b)}']
        
        # Create HORIZONTAL bar plot with BRIGHT COLORS and GRADIENT effect
        bars = ax.barh(categories, percentages, color=color_scheme, alpha=0.95, 
                      edgecolor='white', linewidth=2.5, height=0.5)  # Even thinner bars!
        
        # Add GLOW effect to bars
        for i, bar in enumerate(bars):
            # Add shadow/glow effect
            shadow_bar = ax.barh(categories[i], percentages[i], 
                               color=color_scheme[i], alpha=0.25, 
                               height=0.7, zorder=bars[i].zorder-1)
        
        # Calculate max width for better positioning
        max_width = max(percentages)
        # Add value labels with IMPROVED positioning to avoid overlap
        for i, (bar, sample) in enumerate(zip(bars, sample_info)):
            width = bar.get_width()
            bar_y = bar.get_y() + bar.get_height()/2.
            
            # PERCENTAGE LABEL - Always at the tip/end of the bar (CLEAN TEXT - NO BOX)
            ax.text(width + max_width*0.015, bar_y,
                   f'{width:.1f}%', ha='left', va='center', 
                   fontsize=13, fontweight='bold', color='black')
            
            # SAMPLE SIZE (n) - IMPROVED positioning logic to avoid overlap (CLEAN TEXT - NO BOX)
            if width > 20:  # If bar is wide enough, place inside at 1/3 position
                ax.text(width/3, bar_y,
                       f'{sample}', ha='center', va='center', 
                       fontsize=12, fontweight='bold', color='black')
            elif width > 8:  # Medium width bars - place inside at center
                ax.text(width/2, bar_y,
                       f'{sample}', ha='center', va='center', 
                       fontsize=12, fontweight='bold', color='black')
            else:  # Narrow bars - place outside with more spacing
                ax.text(width + max_width*0.12, bar_y,
                       f'{sample}', ha='left', va='center', 
                       fontsize=12, fontweight='bold', color='black')
        
        # Add significance annotation if p-value available (CLEAN TEXT ONLY - NO BOX)
        if not pd.isna(p_value):
            x_sig = max_width + max_width * 0.3  # Adjusted for better spacing
            
            # Draw MODERN significance bracket
            bracket_color = '#444444'
            ax.plot([x_sig, x_sig], [-0.25, 1.25], color=bracket_color, linewidth=2.5)
            ax.plot([x_sig - max_width*0.015, x_sig], [-0.25, -0.25], color=bracket_color, linewidth=2.5)
            ax.plot([x_sig - max_width*0.015, x_sig], [1.25, 1.25], color=bracket_color, linewidth=2.5)
            
            # Add p-value with CLEAN TEXT (NO BACKGROUND BOX)
            if significance != "ns":
                text_color = '#FF0000' if p_value < 0.01 else '#FF6600'  # Red for highly sig, orange for sig
                ax.text(x_sig + max_width*0.025, 0.5, f'{significance}\np = {p_value:.4f}', 
                       ha='left', va='center', fontsize=13, fontweight='bold', color=text_color)
            else:
                ax.text(x_sig + max_width*0.025, 0.5, f'ns\np = {p_value:.4f}', 
                       ha='left', va='center', fontsize=11, fontweight='bold', color='#666666')
        
        # Plot labels
        ax.set_xlabel('Percentage (%)', fontsize=13, fontweight='bold', color='#1A1A1A')
        ax.set_ylabel('Response Choices', fontsize=13, fontweight='bold', color='#1A1A1A')
        
        # SMART axis range - adjusted for improved label positioning
        ax.set_xlim(0, max_width * 1.7)  # Increased to accommodate new spacing
        
        # MODERN grid
        ax.grid(True, alpha=0.25, axis='x', linestyle=':', color='#AAAAAA')
        ax.set_axisbelow(True)
        
        # CLEAN styling
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['left'].set_linewidth(2)
        ax.spines['left'].set_color('#444444')
        ax.spines['bottom'].set_linewidth(2)
        ax.spines['bottom'].set_color('#444444')
        
        # CLEAN TOTAL INFO (NO BOX - TEXT ONLY)
        ax.text(0.98, 0.02, f'Total: {int(total_responsive)}', 
               transform=ax.transAxes, fontsize=11, fontweight='bold', 
               va='bottom', ha='right', color='#333333')
        
        # Add subtle background gradient effect
        ax.set_facecolor('#FCFCFC')
        
        plt.tight_layout()
        
        # Export individual graph with SMART naming
        safe_filename = f'horizontal_smart_{idx+1}_total{int(total_responsive)}_bright_600dpi'
        if merge_rows_34 and idx == 2:
            safe_filename = f'horizontal_smart_3_4_combined_total{int(total_responsive)}_bright_600dpi'
            
        plt.savefig(f'{safe_filename}.png', 
                    dpi=600, bbox_inches='tight', facecolor='white', format='png')
        plt.savefig(f'{safe_filename}.jpg', 
                    dpi=600, bbox_inches='tight', facecolor='white', format='jpeg')
        
        print(f"‚úÖ Exported: {safe_filename}.png")
        print(f"‚úÖ Exported: {safe_filename}.jpg")
        
        plt.show()
        
        # Add to results summary
        results_summary.append(row_data)
        
        # Print SMART summary
        print(f"\n--- {row_data['row_name']} Summary ---")
        print(f"Total Responsive: {int(total_responsive)}")
        print(f"{choice_a_label}: {percent_a:.1f}% (n={int(n_choice_a)}) - Bright Color 1")
        print(f"{choice_b_label}: {percent_b:.1f}% (n={int(n_choice_b)}) - Bright Color 2")
        if not pd.isna(p_value):
            print(f"P-value: p = {p_value:.4f} {significance}")
        print("-" * 50)
        
        processed_rows.add(idx)
    
    return pd.DataFrame(results_summary)

# Run the analysis with SMART horizontal graphs and BRIGHT colors
print("Creating SMART horizontal percentage comparison graphs with BRIGHT colors...")
results_df = create_horizontal_percentage_graphs_with_p_column(df)

# Final SMART summary
print("\nüéØ SMART ANALYSIS SUMMARY üéØ")
print("=" * 60)
if not results_df.empty:
    for idx, result in results_df.iterrows():
        sig_text = "SIGNIFICANT" if result.get('significance', 'ns') != 'ns' else "NOT SIGNIFICANT"
        print(f"{result['row_name']}: "
              f"Total: {int(result['total_responsive'])}, "
              f"{result.get('choice_a_label', 'Choice A')}: {result['percent_a']:.1f}% (n={int(result['n_choice_a'])}), "
              f"{result.get('choice_b_label', 'Choice B')}: {result['percent_b']:.1f}% (n={int(result['n_choice_b'])}) - {sig_text}")

    print(f"\n‚úÖ Created {len(results_df)} SMART horizontal comparison graphs")
    print("üé® BRIGHT vibrant colors with gradient effects")
    print("üìä THINNER design for better visual appeal")
    print("üß† SMART label positioning (percentage at tip, n inside/outside)")
    print("üî¨ MODERN significance brackets with color coding")
    print("üí´ GLOW effects and subtle background gradients")
    print("üìà CLEAN text without background boxes")
    print("üè∑Ô∏è Dynamic labels from first column")
    print("üîÑ Rows 3 and 4 combined")
    print("üö´ No titles, no overlap issues fixed")
else:
    print("‚ùå No graphs created - check column names")
```

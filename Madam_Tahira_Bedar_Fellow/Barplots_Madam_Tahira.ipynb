```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import os

# Refresh the DataFrame
df = pd.read_excel("C:\\Users\\mirza\\OneDrive\\Desktop\\Phyioslogy data2021 & 2022 19-06-2025.xlsx", sheet_name="Data")
df.rename(columns={'Variety': 'Ecotype'}, inplace=True)

# Output directory
output_dir = "exported_plots_png"
os.makedirs(output_dir, exist_ok=True)

# Get all numeric columns in pairs
numeric_cols = df.select_dtypes(include='number').columns.tolist()

for i in range(0, len(numeric_cols) - 1, 2):
    value_col = numeric_cols[i]
    se_col = numeric_cols[i + 1]

    # Prepare data
    plot_data = df.copy()
    plot_data["Value"] = plot_data[value_col]
    plot_data["SE"] = plot_data[se_col]
    plot_data["Ecotype_Stress"] = plot_data["Ecotype"] + "_" + plot_data["Stress"]

    # Sort properly
    plot_data["Ecotype"] = pd.Categorical(plot_data["Ecotype"], categories=sorted(plot_data["Ecotype"].unique()), ordered=True)
    plot_data["Stress"] = pd.Categorical(plot_data["Stress"], categories=["Control", "Heat"], ordered=True)
    plot_data = plot_data.sort_values(["Ecotype", "Stress"])
    plot_data["Ecotype_Stress"] = pd.Categorical(plot_data["Ecotype_Stress"], categories=plot_data["Ecotype_Stress"].unique(), ordered=True)

    # Plot
    plt.figure(figsize=(12, 6))
    ax = sns.barplot(data=plot_data, x="Ecotype_Stress", y="Value", hue="Treatment", dodge=True, errorbar=None)

    # Add error bars to the tip of each bar
    for bar, row in zip(ax.patches, plot_data.itertuples()):
        bar_height = bar.get_height()  # Get the height of the bar
        bar_center = bar.get_x() + bar.get_width() / 2  # Calculate the center of the bar
        ax.errorbar(bar_center, bar_height, yerr=row.SE, fmt='none', color='black', capsize=3)

    # Adjust X-axis labels
    plt.xlabel("Ecotype\nStress")
    plt.ylabel(value_col)
    plt.xticks(rotation=45, ha='right')  # Rotate labels for better visibility
    plt.tight_layout()

    # Save plot
    filename = os.path.join(output_dir, f"{value_col.replace('/', '_').replace(' ', '_')}.png")
    plt.savefig(filename, dpi=300)
    plt.close()
```

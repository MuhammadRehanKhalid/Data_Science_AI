```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_tables_from_url(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
    }

    response = requests.get(url, headers=headers)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, "html.parser")

    tables = soup.find_all("table")

    if not tables:
        print("No tables found on this page.")
        return

    print(f"Found {len(tables)} tables.\n")

    for i, table in enumerate(tables):
        rows = table.find_all("tr")
        table_data = []

        for row in rows:
            cols = row.find_all(["td", "th"])
            cols = [col.get_text(strip=True) for col in cols]
            if cols:
                table_data.append(cols)

        if not table_data:
            continue

        # Try to use first row as header
        df = pd.DataFrame(table_data)

        filename = f"table_{i+1}.csv"
        df.to_csv(filename, index=False)
        print(f"Saved: {filename}")

    print("\nDone scraping tables.")

# 🔹 ENTER ANY WEBSITE URL HERE
url = "http://timeshighereducation.com/world-university-rankings/latest/world-ranking"

scrape_tables_from_url(url)
```

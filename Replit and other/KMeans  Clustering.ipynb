```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('E:\F1.csv')

# Choose the features for clustering
features = ['V1', 'V2']
X = data[features]

# Run K-means with k=2 clusters
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)

# Assign clusters to the data points
clusters = kmeans.predict(X)

# Visualize the results in a scatter plot
plt.scatter(X['V1'], X['V2'], c=clusters)
plt.title('K-means Clustering Results')
plt.xlabel('V1')
plt.ylabel('V2')
plt.show()

# Run K-means multiple times and store the results
results = []
for i in range(30):
    kmeans = KMeans(n_clusters=2, random_state=i).fit(X)
    clusters = kmeans.predict(X)
    results.append(clusters)

# Visualize the results in separate scatter plots
for i, clusters in enumerate(results):
    plt.subplot(3, 10, i+1)
    plt.scatter(X['V1'], X['V2'], c=clusters)
    plt.title('Run {}'.format(i+2))
    plt.xlabel('V1')
    plt.ylabel('V2')

plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('E:\F1.csv')

# Choose the features for clustering
features = ['V1', 'V2']
X = data[features]

# Run K-means with k=2 clusters
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)

# Assign clusters to the data points
clusters = kmeans.predict(X)

# Visualize the results in a scatter plot
plt.scatter(X['V1'], X['V2'], c=clusters)
plt.title('K-means Clustering Results')
plt.xlabel('V1')
plt.ylabel('V2')
plt.show()

# Run K-means multiple times and store the results
results = []
for i in range(30):
    kmeans = KMeans(n_clusters=3, random_state=i).fit(X)
    clusters = kmeans.predict(X)
    results.append(clusters)

# Visualize the results in separate scatter plots
fig, axs = plt.subplots(nrows=3, ncols=10, figsize=(30, 9))
for i, clusters in enumerate(results):
    ax = axs[int(i / 10)][i % 10]
    ax.scatter(X['V1'], X['V2'], c=clusters)
    ax.set_title('Run {}'.format(i+2))
    ax.set_xlabel('V1')
    ax.set_ylabel('V2')

plt.tight_layout()
plt.show()
```

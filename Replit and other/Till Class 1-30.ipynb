```python
#------------Class28---- Epic Battleground Character-----------------#
#-------------------Class 27 exercise-------------------#
import os, time
import random as rd
from termcolor import colored 
def roll_dice (sides): 
    import random as rd
    dice = rd.randint(1, sides)
    return dice
def health():
    health_stat = ((roll_dice(6)*roll_dice(12))/2)+10
    return health_stat
def strength():
    strength_stat = ((roll_dice(6)*roll_dice(8))/2)+12
    return strength_stat
time.sleep(1)
print("âš”ï¸âš”ï¸âš”ï¸âš”ï¸âš”ï¸ Character Builder âš”ï¸âš”ï¸âš”ï¸âš”ï¸âš”ï¸" )
time.sleep(1)
os.system("clear")
chr_1 = input("Enter your character name: ")
Chr_1_ty = input("Character Type (Human, Elf, Wiard, Orc):")
print(chr_1)
print()
Chr_1_health = health()
Chr_1_strength = strength()
print("Health", Chr_1_health)
print("Strength", Chr_1_strength)
print(Chr_1_ty)
print("To whom are they going to fight?")
print()
chr_2 = input("Enter your character name: ")
Chr_2_ty = input("Character Type (Human, Elf, Wiard, Orc):")
print(chr_2)
print(Chr_2_ty)
Chr_2_health = health()
Chr_2_strength = strength()
print("Health", Chr_2_health)
print("Strength", Chr_2_strength)


round = 1
Winner = None

while True:
    time.sleep(1)
    os.system("clear")
    print("Battle Begins âš”ï¸âš”ï¸")
    Chr_1_d = roll_dice(6)
    Chr_2_d = roll_dice(6)

    diff = abs(Chr_1_d - Chr_2_d)+1
    if chr_1> chr_2:
        Chr_2_health -= diff
        if round == 1:
            print(chr_1, "Won the first blow") 
        else:
            print(chr_1, "Won the round")
    elif chr_2 > chr_1:
        Chr_1_health -= diff
        if round == 1:
            print(chr_2, "Won the first blow") 
        else:
            print(chr_2, "Won the round")
    else:
        print("Draw", round)
    print()
    print(chr_1)
    print("HEALTH:", Chr_1_health)
    print()
    print(chr_2)
    print("HEALTH:", Chr_2_health)
    print()

    if Chr_1_health<=0:
        print(chr_1, "has died!")
        winner = chr_2
        break
    elif Chr_2_health<=0:
        print(chr_2, "has died!")
        winner = chr_1
        break
    else:
        print("And they're both standing for the next round")
        round += 1
time.sleep(1)
os.system("clear")
print("âš”ï¸ BATTLE TIME âš”ï¸")
print()
print(winner, "has won in", round, "rounds")
```

```python
#Data set started titanic
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load Data set
df_tit = sns.load_dataset("titanic")
```

```python
# df_tit.info()
df_tit.isnull().sum()       #count the missing values
df_tit.isnull().sum()/ len(df_tit) * 100
sns.heatmap(df_tit.isnull())
```

```python
df_tit['sex'].unique()
df_tit[['age','sex','fare']] #use double brackets for multiple columns
df_tit.nunique() #unique values
df_tit.columns
df_tit['embark_town'].unique() #find a unique values in a column
#count the number of each value
df_tit['embark_town'].value_counts()
```

```python
df_tit['fare'].mean()
#use group by function to calculate male and female mean fare
df_tit.groupby('sex')['fare'].mean()
df_tit.groupby(['sex','class'])['fare'].mean()
df_tit.groupby(['sex','class','who'])['fare'].size()
df_tit.groupby(['survived','who']).size()
```

```python
corr_df = df_tit[['fare','age','sibsp','parch']].corr()
sns.heatmap(corr_df,annot=True)
```

```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# === 1. Load Image ===
img = cv2.imread("./01_Jeronimo_Lab_Program/IMG-20251031-WA0037.jpg")
scale = 0.8
img = cv2.resize(img, (int(img.shape[1]*scale), int(img.shape[0]*scale)))

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# === 2. Detect the real hemocytometer grid ===
edges = cv2.Canny(gray, 40, 120)
lines = cv2.HoughLinesP(edges, 1, np.pi/180, threshold=120, minLineLength=100, maxLineGap=10)

if lines is not None:
    xs, ys = [], []
    for line in lines:
        x1, y1, x2, y2 = line[0]
        xs += [x1, x2]
        ys += [y1, y2]
    x_min, x_max = int(np.percentile(xs, 10)), int(np.percentile(xs, 90))
    y_min, y_max = int(np.percentile(ys, 10)), int(np.percentile(ys, 90))
else:
    h, w = img.shape[:2]
    x_min, x_max = w//4, 3*w//4
    y_min, y_max = h//4, 3*h//4

# Crop to the real central grid region
grid = img[y_min:y_max, x_min:x_max]

# === 3. Color-based segmentation for viable and dead cells ===
hsv = cv2.cvtColor(grid, cv2.COLOR_BGR2HSV)

# Viable cells (greenish pigment)
lower_green = np.array([30, 25, 25])
upper_green = np.array([95, 255, 255])
mask_green = cv2.inRange(hsv, lower_green, upper_green)

# Dead cells (whitish or opaque)
lab = cv2.cvtColor(grid, cv2.COLOR_BGR2LAB)
L, A, B = cv2.split(lab)
mask_white = cv2.inRange(L, 180, 255)  # bright tones

# Combine both for total
mask_total = cv2.bitwise_or(mask_green, mask_white)

# === 4. Detect contours and count ===
output = grid.copy()
viable_count = 0
dead_count = 0

contours, _ = cv2.findContours(mask_total, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

for cnt in contours:
    area = cv2.contourArea(cnt)
    if 30 < area < 2500:
        x, y, w, h = cv2.boundingRect(cnt)
        cx, cy = x + w//2, y + h//2

        # Apply L-rule (ignore bottom/right edge)
        if cx >= grid.shape[1]-5 or cy >= grid.shape[0]-5:
            continue

        # Determine type based on mask overlap
        roi_green = mask_green[y:y+h, x:x+w]
        green_ratio = np.sum(roi_green > 0) / (w*h)

        if green_ratio > 0.25:
            color = (0, 255, 0)  # green = viable
            viable_count += 1
        else:
            color = (255, 255, 255)  # white = dead
            dead_count += 1

        cv2.rectangle(output, (x, y), (x+w, y+h), color, 1)
        cv2.circle(output, (cx, cy), 4, color, -1)

# === 5. Compute stats ===
total_cells = viable_count + dead_count
viability_ratio = (viable_count / total_cells * 100) if total_cells > 0 else 0

print(f"✅ Viable cells (green): {viable_count}")
print(f"☠️  Dead cells (white): {dead_count}")
print(f"📊 Total: {total_cells}")
print(f"💚 Viability ratio: {viability_ratio:.2f}%")

# === 6. Save & display ===
cv2.putText(output, f"Viable: {viable_count}", (10, 20),
            cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)
cv2.putText(output, f"Dead: {dead_count}", (10, 45),
            cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
cv2.putText(output, f"Viability: {viability_ratio:.1f}%", (10, 70),
            cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 0), 2)

output_rgb = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8, 8))
plt.imshow(output_rgb)
plt.title("Hemocytometer Cell Count: Viable vs Dead")
plt.axis("off")
plt.show()

# Save results
cv2.imwrite("hemocytometer_result.jpg", output)
pd.DataFrame({
    "Viable": [viable_count],
    "Dead": [dead_count],
    "Total": [total_cells],
    "Viability_%": [viability_ratio]
}).to_csv("hemocytometer_viability.csv", index=False)
```

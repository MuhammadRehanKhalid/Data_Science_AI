```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['pCG', 'pC5S', 'C5S', 'SM', 'K3R7R']
groups = ['WT', 'hct D7', 'Î”HCT', 'PpHCT', 'MpHCT']
values = np.random.rand(5, 5)  # Replace with your actual data
errors = np.random.rand(5, 5) * 0.1  # Replace with your actual error data
significance = ['***', '**', '*', 'ns', 'ns']  # Replace with your actual significance data

x = np.arange(len(categories))
width = 0.15

fig, ax = plt.subplots()

for i, group in enumerate(groups):
    ax.bar(x + i*width, values[:, i], width, label=group, yerr=errors[:, i], capsize=5)

# Adding significance annotations
for i, (cat, sig) in enumerate(zip(categories, significance)):
    ax.text(i, max(values[i]) + max(errors[i]) + 0.05, sig, ha='center')

ax.set_xlabel('Category')
ax.set_ylabel('Metabolite level (relative to WT)')
ax.set_xticks(x + width * 2)
ax.set_xticklabels(categories)
ax.legend()

plt.show()
```

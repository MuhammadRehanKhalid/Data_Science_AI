```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate data 
df = np.random.normal(100, 20, 200)
df = np.append(df, [300, 2])
```

```python
# Calculate qauartile
Q1 = np.percentile(df, 25)
Q3 = np.percentile(df, 75)
IQR = Q3 - Q1
print(f"Q1 = {Q1}\nQ3 = {Q3}\nIQR = {IQR}")
```

```python
# find upper and lower bound
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR
print(f"upper_bound = {upper_bound}\nlower_bound = {lower_bound}")
```

```python
# identify the outliers in data
outliers = (df < lower_bound) | (df > upper_bound)
print(f"outliers = {outliers}")
```

```python
# remove outliers
df_nu = df[(df >= lower_bound) & (df <= upper_bound)]
```

```python
# plot the original data with outliers and the final results
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
sns.boxplot(df, color="red")
#
plt.subplot(1,2,2)
sns.boxplot(df_nu, color="green")
plt.show()
```

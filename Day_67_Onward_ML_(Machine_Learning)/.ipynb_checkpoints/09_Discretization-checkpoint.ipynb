#### Discretization is the process of converting continuous data into discrete categories or intervals. 
#### It's often used to prepare data for numerical evaluation and implementation on digital computers

```python
# making data bins
# pandas, scikit learn for discretization
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import KBinsDiscretizer
```

```python
df = sns.load_dataset('titanic')
```

```python
# impute the missing values
df['age'] = df['age'].fillna(df['age'].median())
df['fare'] = df['fare'].fillna(df['fare'].median())
sns.histplot(df['age'], kde=True)
```

```python
df_2 = df.drop(columns=['deck'])
df_2
```

```python
figure, axs = plt.subplots(2,3,figsize=(20, 10))
age_desc = KBinsDiscretizer(n_bins=2, encode='ordinal', strategy='kmeans') 
df_2['age_bin'] = age_desc.fit_transform(df_2[['age']])
df_2.head()
sns.histplot(df_2, x = 'age', kde=True, hue = 'age_bin', ax = axs[0,0])
age_desc_1 = KBinsDiscretizer(n_bins=2, encode='ordinal', strategy='uniform') 
df_2['age_bin'] = age_desc_1.fit_transform(df_2[['age']])
df_2.head()
sns.histplot(df_2, x = 'age', kde=True, hue = 'age_bin', ax = axs[0,1])
age_desc_1 = KBinsDiscretizer(n_bins=2, encode='ordinal', strategy='quantile') 
df_2['age_bin'] = age_desc_1.fit_transform(df_2[['age']])
df_2.head()
sns.histplot(df_2, x = 'age', kde=True, hue = 'age_bin', ax = axs[0,2])
age_desc_1 = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='kmeans') 
df_2['age_bin'] = age_desc_1.fit_transform(df_2[['age']])
df_2.head()
sns.histplot(df_2, x = 'age', kde=True, hue = 'age_bin', ax = axs[1,0])
age_desc_1 = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='uniform') 
df_2['age_bin'] = age_desc_1.fit_transform(df_2[['age']])
df_2.head()
sns.histplot(df_2, x = 'age', kde=True, hue = 'age_bin', ax = axs[1,1])
age_desc_1 = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='quantile') 
df_2['age_bin'] = age_desc_1.fit_transform(df_2[['age']])
df_2.head()
sns.histplot(df_2, x = 'age', kde=True, hue = 'age_bin', ax = axs[1,2])
```

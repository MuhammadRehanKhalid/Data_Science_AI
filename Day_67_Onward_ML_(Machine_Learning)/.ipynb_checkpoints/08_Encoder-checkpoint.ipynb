- OneHot Encoding

- Binary Encoding (onehot  encoding comes here also)

### Benefits of Encoding
- Algorithem Compatibility
- To convert text data into numeric data
- efficient storage (occupy less space)
- faster computation
- 

```python
# Feature Encoding by scikitlearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import data
df = sns.load_dataset('tips')
df.head()
```

```python
print(df['time'].value_counts())
print(df['day'].value_counts())
print(df['time'].value_counts())
```

```python
# use sklearn
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
```

```python
le = LabelEncoder()
df['encoded_time'] = le.fit_transform(df['time'])
df['encoded_time'].value_counts() # dinner  = 0, lunch is = 1
df
```

```python
# ordinal encoding
Oe = OrdinalEncoder(categories=[['Thur','Fri','Sat','Sun']])
df['encoded_day'] = Oe.fit_transform(df[['day']])
df['encoded_day'].value_counts() # dinner  = 0, lunch is = 1
df.drop(columns=['encoded_days'], inplace=True)
df
```

```python
# ordinal encoding
One = OneHotEncoder()
One.fit_transform(df[['day']]).toarray()
```
